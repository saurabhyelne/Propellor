### Perform following steps on US Judge Ratings data <br>                                                                                          Task: Perform principal component analysis<br>
• Input the raw data matrix to fa.parallel() function to determine the number of components to extract<br>
• Input the raw data matrix to principal() function to extract the components. If raw data is input, the correlation     matrix is automatically calculated by principal() function.<br>
• Rotate the components<br>
• Compute component scores<br>
• Graph an orthogonal solution using factor.plot()<br>
• Interpret the results

```r
head(USJudgeRatings)
usj<-USJudgeRatings[,-1]
corr<-cor(usj)
```

## 1.Input the raw data matrix to fa.parallel() function to determine the number of components to extract

```r
fa.parallel(corr, fa = 'pc',n.iter = 100,n.obs=43,
            main='Scree plot for US Judge Ratings with parallel analysis')
+abline(h=1)
```

### As only one component has eigenvector value > 1, we should extract 1 component

## 2.Input the raw data matrix to principal() function to extract the components. If raw data is input, the correlation matrix is automatically calculated by principal() function.

```r
pc<-principal(USJudgeRatings[,-1],nfactors = 1)
```

```r
pc
```

## 3.Rotate the components

```r
pcpro<-principal(USJudgeRatings[,-1],nfactors = 1,rotate = 'promax')
```

```r
pcvari<-principal(USJudgeRatings[,-1],nfactors = 1,rotate = 'varimax')
```

```r
pcpro
```

```r
pcvari
```

### The components remains the same when rotated orthogonally or oblique

## 4.Compute component scores

```r
head(pc$scores,10)
```

## 5.Graph an orthogonal solution using factor.plot()

```r
factor.plot(pcvari,labels = rownames(pcvari$loadings), title ='Orthogonal Solution for US Judge ratings')
```

## 6.Interpret the results

1.Based on the Scree plot with parallel analysis and Kaiser-Harris (eigenvalue greater than 1) criteria, the suggested number of components to extract is ONE. Using the principal () to extract the components the result yields components that are highly correlated to PC1.<br><br>
2.The column h2 represents communalities components and u2 represents the uniqueness component and they are related by the equation u2=(1-h2). <br><br>
3.Quantity SS is the eigenvalue associated with PC1 and it accounts for 92% of the variance in all of the 11 variables. When we compute the principal component using orthogonal rotation or rotate=”varimax”where the purpose is to purify the components or column of the loading matrix, we found no difference in the result without rotation and the result with rotation.
This is probably due only one component being selected. The result should not be the same when we have more than two components selected.<br><br> 4.When we calculated the principal component scores and loading factor for each of the 11 lawyers, we again see that they are highly correlated to PC1 where the numeric values range from 0.894 to 0.995 likewise as shown by the Principal Component Analysis plot of index versus load factor.

## Perform following steps on Vertebral Column Data.xlsx <br>Task: Perform multidimensional scaling
• Input the raw data matrix to fa.parallel() function to determine the number of components to extract<br>
• Input the raw data matrix to cmdscale() function to perform multidimensional scaling. cmdscale() function which is available in the base installation performs a classical multidimensional scaling.<br>
• Graph an orthogonal solution using factor.plot()<br>
• Interpret the results

```r
str(Vertebral_Column_Data)
vertebral<-Vertebral_Column_Data[,-7]
head(vertebral)
```

## 1.Input the raw data matrix to fa.parallel() function to determine the number of components to extract

```r
fa.parallel(vertebral,n.obs = 310,n.iter = 100,fa = 'both',
            main = 'Scree plot for Vertebral_Column_Data with parallel analysis ')+abline(h=0)
```

### The number of principal components extracted should be 2 and number of factors extracted are 3

## 2.Input the raw data matrix to cmdscale() function to perform multidimensional scaling. cmdscale() function which is available in the base installation performs a classical multidimensional scaling.

```r
disv<-dist(vertebral,method ='euclidean',p=2,diag = F,upper = F)
```

```r
disv
```

```r
mds<-cmdscale(disv,k=2)
```

```r
mds
```

## 3.Graph an orthogonal solution using factor.plot()

```r
factor.plot(mds,labels = "")
```

## 4.Interpret the results

1.The dist Function Computes the distance matrix by euclidean distance. Then this distance matrix is feed in cmdscale function for dimension reduction, where dimensions are reduced from 6 variables to 2 variables<br><br>2. The number of principal components extracted should be 2 and number of factors extracted are 3
<br><br>3.The dataset vertebral columns has 6 attributes and 3 disease classes. By applying function fa.parallel(), it suggests 2 principal components and 3 factors should be included in the analysis.<br><br> 4.By putting raw data into dist(), the function calculates the Euclidean distance that represents similarity in a symmetric matrix, which is used to conduct multidimensional scaling. In the outcome plot, 310 data points have been graphed on the two-dimensional plane, which try to reproduces the observed distances computed previously to the greatest extent. We can clearly see three clusters on the plot that represent three classes of patients.
